```python
#On supprime les valeurs manquantes
data=data.dropna(subset=["artists", "album_name", "track_name"])
data=data.drop(columns=["Unnamed: 0"])
```

```python
data.head()
```

```python
feature=["danceability", "energy", "loudness", "speechiness", "acousticness",
    "instrumentalness", "liveness", "valence", "tempo"]

data_feature= data[feature]
```

```python

from sklearn.preprocessing import StandardScaler

scaler= StandardScaler()
feature_scaled= scaler.fit_transform(data_feature)
```

```python
from sklearn.neighbors import NearestNeighbors
model= NearestNeighbors(n_neighbors=6, metric='cosine')
model.fit(feature_scaled)
```

```python
def recommend(song_name, df_data, model, features_scaled_data):
    # Rendre la recherche insensible à la casse
    matches = df_data[df_data['track_name'].str.lower() == song_name.lower()]
    
    if matches.empty:
        return f"❌ La chanson '{song_name}' n'existe pas dans le dataset."
    
    # On prend le premier match trouvé (au cas où il y aurait plusieurs)
    idx = matches.index[0]

    # Trouver les voisins les plus proches
    distances, indices = model.kneighbors([features_scaled_data[idx]])

    # Construire la liste de recommandations
    recommendations = []
    for i in indices[0][1:]:  # On saute la 1ère (la chanson elle-même)
        track = df_data.iloc[i]
        recommendations.append({
            "track_name": track["track_name"],
            "artist": track["artists"]
        })

    return recommendations
```

```python
results = recommend("Hold On", data, model, feature_scaled)

for i, rec in enumerate(results, 1):
    print(f"{i}. {rec['track_name']} — {rec['artist']}")
```
