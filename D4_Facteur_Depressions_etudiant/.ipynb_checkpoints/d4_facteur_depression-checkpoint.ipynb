```python
import numpy as np
import pandas as pd
import time
from mlxtend.frequent_patterns import apriori, association_rules
from collections import defaultdict
```

```python
data= pd.read_csv('student_depression_dataset.csv')
```

```python
#On verifie les valeurs unique de la colonne Dietary Habits pour savoir si on le garde ou pas
valeurs_uniques = data['Dietary Habits'].unique()
print(valeurs_uniques)
```

```python
#Supprimons les colonnes inutiles
drop_colo=['id', 'City', 'Profession','Work Pressure', 'Degree', 'Job Satisfaction']

data_clean= data.drop(columns=drop_colo, errors= 'ignore')
```

```python
data_clean.head()
```

```python
#Discrétisons les variables numériques
#Variable Age
data_clean['Age']=pd.qcut(data['Age'], q=4, duplicates= 'drop', labels=['Very Low', 'Low', 'Hight', 'Very Hight']).astype(str)

#Variable Academic Pressure
data_clean['Academic Pressure']=pd.qcut(data['Academic Pressure'], q=4, duplicates= 'drop', labels=['Very Low', 'Low', 'Hight', 'Very Hight']).astype(str)

#Variable CGPA
data_clean['CGPA']=pd.qcut(data['CGPA'], q=4, duplicates= 'drop', labels=['Very Low', 'Low', 'Hight', 'Very Hight']).astype(str)

#Variable CGPA
data_clean['Study Satisfaction']=pd.qcut(data['Study Satisfaction'], q=4, duplicates= 'drop', labels=['Very Low', 'Low', 'Hight', 'Very Hight']).astype(str)

```

```python
data_clean.head()
```

```python
#Convertissons chaque ligne en semble d'itemset
transactions = data_clean.astype(str).apply(lambda x: set(x.dropna()), axis=1).tolist()
print(transactions[:5])
```

```python
# Créer un dictionnaire pour stocker tous les items uniques
all_items = defaultdict(int)
for transaction in transactions:
    for item in transaction:
        all_items[item] += 1
```

```python
# Créer un DataFrame binaire
df_trans = pd.DataFrame([[item in transaction for item in all_items.keys()] for transaction in transactions], columns=all_items.keys())
```

```python
# Mesurer le temps d'exécution
start_time = time.time()

# Utilisation de l'algo apriori
frequent_itemsets = apriori(df_trans, min_support=0.01, use_colnames=True)

# Génération des règles d'association
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)

# Affichage des règles triées par confiance (ordre décroissant)
rules = rules.sort_values(by="confidence", ascending=False)

apriori_time = time.time() - start_time


```

```python
# Affichage des premières règles générées
print(rules.head())
print(f"Apriori terminé en {apriori_time:.4f} secondes")

# Enregistrement des résultats pour analyse ultérieure
rules.to_csv("apriori_rules.csv", index=False)
print("Règles sauvegardées sous 'apriori_rules.csv' !")
```

Règles d’association qui ont comme conséquence la variable cible Depression.

```python
rules= pd.read_csv("apriori_rules.csv")
# Filtrer les règles où y est le conséquent
rules_target = rules[rules['consequents'].astype(str).str.contains("'1'")]

# Trier par confiance (ordre décroissant)
rules_target = rules_target.sort_values(by="confidence", ascending=False)

# Afficher les 5 premières règles trouvées
print("Règles qui prédisent Dépression = 1 :")
print(rules_target[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head())

# Sauvegarde des résultats
rules_target.to_csv("apriori_rules_target.csv", index=False)
print("Règles sauvegardées sous 'apriori_rules_target.csv' !")
```
